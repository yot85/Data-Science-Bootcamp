```python
import pandas as pd
import numpy as np
```

```python
url = ('https://storage.googleapis.com/esmartdata-courses-files/ds-bootcamp/online_retail.xlsx')
reatil_raw = pd.read_excel(url)
reatil_raw.head()
```

```python
retail = reatil_raw.copy()
```

```python
retail.info()
```

```python
retail.describe().T
```

```python
retail.isnull().sum()
```

```python
retail = retail.dropna()
retail.isnull().sum()
```

```python
retail = retail[retail.Quantity >= 0]
```

```python
retail.describe().T
```

```python
retail.CustomerID = retail.CustomerID.apply(lambda x: str(int(x)))
retail.info()
```
