```python
import matplotlib
matplotlib.__version__
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
fig.show()
```

```python
_ = plt.plot([0, 1], [0, 3])
```

```python
plt.plot([0, 1], [0, 3], label ='line', color ='green')
plt.xlabel('oś X')
plt.ylabel('oś Y')
plt.title('Wykres')
plt.legend()
```

```python
plt.plot([0, 1, 2], [0, 3, 2], label ='line', color ='green')
plt.xlabel('oś X')
plt.ylabel('oś Y')
plt.title('Wykres')
plt.legend()
```

```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
plt.plot(x, x_cos)
```
