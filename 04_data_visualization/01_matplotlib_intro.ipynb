```python
import matplotlib
matplotlib.__version__
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
fig.show()
```

```python
_ = plt.plot([0, 1], [0, 3])
```

```python
plt.plot([0, 1], [0, 3], label ='line', color ='green')
plt.xlabel('oś X')
plt.ylabel('oś Y')
plt.title('Wykres')
plt.legend()
```

```python
plt.plot([0, 1, 2], [0, 3, 2], label ='line', color ='green')
plt.xlabel('oś X')
plt.ylabel('oś Y')
plt.title('Wykres')
plt.legend()
```

```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
plt.plot(x, x_cos)
```

Tworzenie wielu wykresów

```python
plt.subplot(121)
plt.plot([0, 1, 2], [3, 4, 5], 'b')
plt.subplot(122)
plt.plot([2,3,4], [0, 7,2], 'r--')
```

```python
fig = plt.figure(figsize=(12, 12))

plt.subplot(131)
plt.plot([0, 1, 2], [3, 4, 5])
plt.subplot(132)
plt.plot([2,3,4], [0, 7,2], 'r--')
plt.subplot(133)
plt.plot([9,1,5], [4, 1, 6], 'ro')

fig.show()
```

WYkresy słupkowe

```python
plt.bar(x=[0, 1, 2, 3, 4], height=[0.1, 1.2, 3.3, 4.5, 7.2])
```

```python
plt.bar(x=['a','b','c','d','e'], height=[0.1, 1.2, 3.3, 4.5, 7.2])
```

```python
plt.barh(y=['a','b','c','d','e'], width=[0.1, 1.2, 3.3, 4.5, 7.2])
```

Wykresy punktowe

```python
x1 = np.random.randn(300)
x2 = np.random.randn(300)
x3 = np.random.randn(300) * 50


plt.scatter(x1, x2, s=x3)
```

```python
x1 = np.random.randn(300)
x2 = np.random.randn(300)
x3 = np.random.randn(300) * 50


plt.scatter(x1, x2, s=x3, alpha = 0.5)
```
